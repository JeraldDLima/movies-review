```python
genre_votes=pd.concat([movies.iloc[:,11:14],movies.iloc[:,16:60]],axis=1)
genre_votes
```

```python
genre_votes["cnt"]=1
genre_votes
```

```python
df_by_g1=genre_votes.groupby("genre_1").agg(np.sum)
df_by_g1
```

```python
df_by_g2=genre_votes.groupby("genre_2").agg(np.sum)
df_by_g2
```

```python
df_by_g3=genre_votes.groupby("genre_3").agg(np.sum)
df_by_g3
```

```python
df_add=df_by_g1.add(df_by_g2,fill_value=0)
df_add=df_add.add(df_by_g3,fill_value=0)
df_add
```

```python
df_genre_10=df_add[df_add['cnt']>=10]
df_genre_10
```

```python
df_genre_10.iloc[:,:-1]=df_genre_10.iloc[:,:-1].divide(df_genre_10['cnt'],axis=0)
df_genre_10
```
